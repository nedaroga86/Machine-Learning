```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([-3,-2,-1,0,1,2,3])
x2 = np.array([9,4,1,0,1,4,9])
```

## y = x^2

```python
plt.scatter(x1,x2,c='r') # Show the distribution
plt.xlabel("X1")
plt.ylabel("X2")
```

## Example 2

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
```

```python
plt.plot(x,y,'r--P',label='y=2x') ## show the relation between X and Y
plt.legend()
plt.title("This is a demo from Plot")
plt.xlabel("X")
plt.ylabel("Y")
```

## Example 3

```python
x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]
```

```python
plt.plot(x1,y1,'b:D',label='y=x^2') ## show the relation between X and Y
plt.legend()
plt.title("This is a demo from plot")
plt.xlabel("X1")
plt.ylabel("Y1")
```

```python
plt.plot(x,y,'r--P',label='y=2x') ## show the relation between X and Y
plt.plot(x1,y1,'b:D',label='y=x^2') ## show the relation between X and Y
```

```python
plt.subplot(211) ## First number 2 rows, second number columns, and third is the position of the current chart
plt.plot(x,y,'r--P', label = 'y=2x')
plt.legend()

plt.subplot(212)
plt.plot(x1,y1,'b:D',label='y=x^2')
plt.legend()

plt.savefig('my_plot.png')
```

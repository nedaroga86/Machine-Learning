```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame(300*np.random.rand(50,4), columns = ['a','b','c','d'])
df.head()
```

```python
df.info()
```

```python
result = df.plot.scatter(x='a', y='b', color='red', label= 'Group 1')
df.plot.scatter(x='c', y='d', color='blue', label= 'Group 2', ax=result)
```

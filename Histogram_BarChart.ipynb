```python
mu = 100 # Equal to the mean
sigma = 15 ## Standar Desviation
x= mu + sigma * np.random.randn(10000) ## Randon Population
```

```python
num_bins = 20
plt.hist(x, bins =num_bins, edgecolor = 'black', facecolor = 'green', density = True)
```

```python
from scipy.stats import norm
```

```python
n, bins, patches = plt.hist(x, bins =num_bins, edgecolor = 'black', facecolor = 'green', density = True)
y = norm.pdf(bins,mu, sigma) ## Calculation of probability function
plt.plot(bins,y,'r--')

plt.xlabel('Numbers')
plt.ylabel('Probabilities')
plt.title('Histogram Demos with Normal distribution')
plt.show()
```

## Example 1

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(0,1,num=500)
y = np.linspace(0,1,num=500)
```

```python
X,Y = np.meshgrid(x,y)
Z = X+Y
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection = '3d')
s = ax.plot_surface(X,Y,Z, cmap=plt.cm.rainbow)
cset = ax.contour(X,Y,Z, zdir = 'z', offset=0, cmap= plt.cm.rainbow)
fig.colorbar(s)
```

```python
x = np.arange(-10,10,0.5)
y = np.arange(-10,10,0.5)
```

```python
X,Y = np.meshgrid(x,y)
Z = X**2+Y**2
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection = '3d')
s = ax.plot_surface(X,Y,Z, cmap=plt.cm.coolwarm)
cset = ax.contour(X,Y,Z, zdir = 'z', offset=0, cmap= plt.cm.coolwarm)
fig.colorbar(s)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
l = ["Python", "java", "C++"]
s = [200, 150, 100]
c = ['green', 'red', 'yellow']
```

```python
plt.pie(s,labels=l, colors = c, autopct ='%1.1f%%')
```

```python
plt.pie(s,labels=l, colors = c, autopct ='%1.1f%%', startangle = 120, explode=(0.1,0,0), shadow = True)
```

```python
s = pd.Series(3*np.random.rand(3), index =['a','b','c']) ## Series should be only index
s
```

```python
s.plot.pie(colors = c, autopct ='%1.1f%%', startangle = 120, explode=(0.1,0,0), shadow = True)
```

```python
df = pd.DataFrame(3*np.random.rand(6,3), index = ['a','b','c','d','e','f'],columns = ['x','y','z'])
df
```

```python
df.plot.pie(subplots=True, figsize=(20, 10));
```
